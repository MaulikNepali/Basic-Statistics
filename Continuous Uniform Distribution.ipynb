# Continuous uniform distribution is one that assigns equal probabilities on each event in the interval

```python
import numpy as np
import pandas as pd
from scipy.stats import binom, poisson, uniform
from matplotlib import pyplot as plt
```

```python
x = uniform.rvs(loc  = 1, scale = 10, size = 10000)
```

```python
plt.hist(x, bins = 20, edgecolor = 'black', color = 'green')
plt.xlabel('value')
plt.ylabel('Frequency of the value')
plt.title('Histogram of U(1, 11) with 10000 samples')
plt.show()
```

```python
plt.hist(x, bins = 20, edgecolor = 'black', density = True, color = 'green') # if density is true, it shows proportions
plt.xlabel('value')
plt.ylabel('Density (Relative Frequency) of the value')
plt.title('Histogram of U(1, 11) with 10000 samples (with density)')
plt.show()
```
