```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
from scipy.stats import norm
```

```python
x = np.arange(-5, 5, 0.000001)
```

```python
y = norm.pdf(x, 1, 2)
```

```python
plt.plot(x, y, color = 'green')
plt.show()
```
